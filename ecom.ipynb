```python
import pandas as pd
```

```python
customers  = pd.read_csv('customers.csv')
orders     = pd.read_csv('orders.csv')
order_items= pd.read_csv('order_items.csv')
products   = pd.read_csv('products.csv')
categories = pd.read_csv('categories.csv')
shipments  = pd.read_csv('shipments.csv')
```

```python
print(customers.head)
```

```python
print(customers.isnull().sum())
```

```python
customers['referral_id'] = customers['referral_id'].fillna(0).astype(int)
```

```python
print(orders.isnull().sum())
```

```python
print(orders.isnull().sum())
```

```python
print(order_items.head())
```

```python
print(order_items.isnull().sum())
```

```python
print(categories.isnull().sum())
```

```python
categories['parent_id']=categories['parent_id'].fillna(0,inplace=True)
```

```python
print(categories.isnull().sum())
```

```python
print(products.isnull().sum())
```

```python
print(shipments.isnull().sum())
```

```python
print(shipments.tail())
```

```python
shipments['delivery_date'] = shipments['delivery_date'].replace('', None)
```

```python
print(shipments.isnull().sum())
```

```python
customers['signup_date']  = pd.to_datetime(customers['signup_date'])
orders['order_date']       = pd.to_datetime(orders['order_date'])
shipments['shipped_date']  = pd.to_datetime(shipments['shipped_date'])
shipments['delivery_date'] = pd.to_datetime(shipments['delivery_date'], errors='coerce')

products['price']        = products['price'].astype(float)
order_items['unit_price'] = order_items['unit_price'].astype(float)
```

```python
print(orders.head)
```

```python
print(orders.duplicated(subset=['order_id']).sum())
```

```python
import os
os.makedirs('cleaned', exist_ok=True)

for name, df in {
    'customers': customers, 'orders': orders,
    'order_items': order_items, 'products': products,
    'categories': categories, 'shipments': shipments
}.items():
    df.to_csv(f'cleaned/{name}.csv', index=False)
    print(f'{name}: {len(df)} rows saved')
```

```python
from sqlalchemy import create_engine
import pandas as pd

# Change user/password to yours
engine = create_engine('mysql+mysqlconnector://root:2007@localhost/ecommerce')

tables = ['categories','products','customers','orders','order_items','shipments']

for t in tables:
    df = pd.read_csv(f'cleaned/{t}.csv')
    df.to_sql(t, engine, if_exists='append', index=False)
    print(f'Loaded {t}: {len(df)} rows')
```
